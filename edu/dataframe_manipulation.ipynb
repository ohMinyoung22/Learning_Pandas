##### DataFrame 조작
DataFrame 타입에 내재된 기능을 이용하여 데이터를 조작하거나 추출할 수 있다.

```python
titanic_dataFrame['survived']
# DataFrame 이름['열 이름'] : 선택한 열의 값을 각 행마다 나타낸다.
# 예시의 경우 'survived' 열의 값들을 각 행마다 가져온 것이다.
# 아래를 보면 3번 index를 가진 행은 survived 열의 값으로 0을 가지고, 1번 index를 가진 행은 survived 열의 값으로 1을 가지는 것을 볼 수 있다.
```

```python
titanic_dataFrame.loc[3]
# DataFrame 이름.loc[index] : index에 해당하는 행의 정보를 가져온다.
# 위 예시에서는 index가 3인 행의 정보를 가져왔다
```

```python
titanic_dataFrame['survived'].value_counts()
# DataFrame 이름['열 이름'].value_counts() : 선택한 열의 값 종류를 기준으로, 값 종류마다 몇 개의 행이 해당되는지 개수를 보여준다.
# 예시의 경우 'survived' 열은 0과 1 두개의 값을 가지고 있으므로, 0과 1 두 종류를 기준으로 분류된다.
# 'survived' 열의 값이 0인 행이 809개, 1인 행이 500개인 것을 확인할 수 있다.
# value_counts()는 시각화를 할 때도 유용하게 쓰이므로 중요하다.
```

```python
titanic_dataFrame['survived'].value_counts().plot(kind='bar')
# plot(kind='bar') : index를 x축으로, index에 해당하는 값을 y축으로 하는 바 차트를 그린다.
# 예시에서는 index가 0, 1 두 종류이고, index에 해당하는 값이 각각 809, 500 이므로 아래와 같이 차트가 작성된다.
```

```python
survived = titanic_dataFrame[titanic_dataFrame['survived'] == 1]['sex'].value_counts()
survived
```

```python
dead = titanic_dataFrame[titanic_dataFrame['survived'] == 0]['sex'].value_counts()
dead
```
