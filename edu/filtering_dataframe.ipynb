```python
import pandas as pd
```

```python
titanic_dataFrame = pd.read_csv("data/titanic.csv")
```

##### Series
Series는 DataFrame의 부분으로 이해할 수 있다.
"열이 하나인 데이터 프레임"을 Series라고 생각하면 된다.

```python
titanic_dataFrame["survived"]
# 위 코드는 DataFrame에서 "survived" 열을 이용하는 Series를 반환한다.
# 즉 오직 "survived" 열만 있는 DataFrame을 반환하는 것이다.
# Series는 (index, 행의 값)의 나열로 구성된다.
```

##### 조건을 만족하는 행 추출하기
DataFrame에서 특정 조건을 만족하는 행만 DataFrame에서 추출할 수 있다.

```python
titanic_dataFrame[titanic_dataFrame["survived"] == 1]
# DataFrame 이름[조건식] 을 통해 조건식을 만족하는 행들로 이루어진 DataFrame을 추출할 수 있다.
# 위 예시에서는 titanic_dataFrame["survived"], 즉 각 행의 survived 열 값과 1을 비교하여 같은 경우 추출하고, 다르면 추출하지 않는다.
# 이렇게 모든 행을 대상으로 비교를 진행하면 조건식을 만족하는 행들만 남아, 새로운 DateFrame이 반환된다.
```

```python
titanic_dataFrame[(titanic_dataFrame["sibsp"] > 1) & (titanic_dataFrame["sibsp"] < 4)]
# 위 예시에서는 각 행의 sibsp 열 값이 1보다 크고 4보다 작은지 비교한다.
# 만약 1과 4 사이라면 해당 행은 추출하여 결과적으로 새로운 DataFrame을 반환한다.
# 이렇게 조건식을 이용하여 DataFrame에서 원하는 값이나 범위를 가진 자료만 추출할 수 있다.
```
