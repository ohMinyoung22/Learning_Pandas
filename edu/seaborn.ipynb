```python
import pandas as pd
titanic = pd.read_csv("data/titanic.csv")
```

##### Seaborn 임포트
데이터 시각화 패키지인 Seaborn을 사용하기 위해 import를 통해 가져온다.

```python
import seaborn as sns
```

```python
titanic.info()
```

##### Bar plot
seaborn의 barplot() 함수를 이용하면 막대그래프를 그릴 수 있다.

```python
sns.barplot(x='sex', y='survived', data=titanic, hue='pclass')
# x 값은 가로축의 변수를 설정한다.
# y 값은 세로축의 변수를 의미한다.
# data는 이 그래프를 그리는 데 사용될 데이터, 즉 DataFrame을 의미한다.
# hue는 x축 변수에 대해 제3의 카테고리 변수를 도입하여 시각화 해준다. 
# 위의 예시의 경우에는 성별 별로 생존 비율을 보여주되, 성별 안에서도 pclass에 따라 분포를 보여준다.
```

##### Box plot
seaborn의 boxplot() 함수를 이용하면 막대그래프의 경우와 비슷하게 박스 플롯을 그릴 수 있다.

```python
csv_2010 = pd.read_csv("data/csv_in_2010.csv")
csv_2010
```

```python
sns.boxplot(x='남편직업분류코드', y='남편연령', data=csv_2010)
# x 값은 가로축의 변수를 설정한다.
# y 값은 세로축의 변수를 의미한다.
# data는 이 그래프를 그리는 데 사용될 데이터, 즉 DataFrame을 의미한다.
```

##### 표 관련 기능

```python
import matplotlib.pyplot as plt
sns.boxplot(x='남편직업분류코드', y='남편연령', data=csv_2010)
plt.ylim(25, 60)
# ylim(min, max) 기능을 이용하여 그래프의 y축 값의 범위를 제한할 수 있다.
# 이를 이용하기 위해서는 matplotlib.pyplot을 import 해야 한다.
```
