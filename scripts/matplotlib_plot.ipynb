### figure

```python
import numpy as np
nums = np.arange(5)
```

```python
plt.plot(nums, nums)
plt.plot(nums, nums*nums)
plt.plot(nums, nums**3)
```

##### figure : axes를 그릴 수 있는 캔버스 (axes가 여러 개 들어갈 수도 있음)
##### axes : 개별 플롯

```python
plt.figure() # 새 figure -> 이후 추가 되는 모든 axes는 이 figure에 추가됨
plt.plot(nums, nums)
plt.figure()
plt.plot(nums, nums*nums)
plt.figure()
plt.plot(nums, nums**3)
```

### title, label

```python
salaries = [55000, 22000, 45500, 40000, 50000]
ages = [20, 25, 30, 32, 45]

plt.plot(ages, salaries)
plt.title('Company salaries')
plt.xlabel("Age", labelpad=10)
plt.ylabel("salaries", labelpad= 30, rotation="horizontal")
```

```python
salaries = [55000, 22000, 45500, 40000, 50000]
ages = [20, 25, 30, 32, 45]

plt.plot(ages, salaries)
plt.title('Company salaries')

plt.xlabel("Age", labelpad=10)
plt.ylabel("salaries", labelpad= 30, rotation="horizontal")

plt.xticks([20, 25, 30, 32, 45], labels=['A', 'B', 'C', 'D', 'E'])

plt.xlim(20, 45)
```

### 범례 추가

```python
plt.figure(figsize=(8, 8))
plt.plot(nums, color="teal", label="linear")
plt.plot(nums**2, color="olive", label="squared")
plt.plot(nums**3, color="violet", label="cubed")

plt.legend(loc='center left')
```

### barplot

```python
plants = ['Spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10, 25, 5, 30, 21]
germinated = [74, 88, 56, 69, 59]
```

```python
plt.figure(figsize=(6, 6))
plt.bar(x=plants, height=germinated, width=0.5, color="seagreen")
plt.title("Germination Rate")
plt.ylabel("Number of Seedlings")
```

```python
plt.figure(figsize=(6, 6))
plt.bar(plants, died, width=0.5, color="red")
plt.title("Failure Rate")
plt.ylabel("Number of Seedlings")
```

```python
plt.figure(figsize=(6, 6))
plt.bar(x=plants, height=germinated, width=-0.3, color="seagreen", align='edge')
plt.bar(plants, died, width=0.3, color="red", align='edge')
plt.title("Germination Rate")
plt.ylabel("Number of Seedlings")
```
